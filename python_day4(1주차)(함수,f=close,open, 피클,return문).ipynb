# 리스트 문자열 파일에 저장하는 법!, 그리고 읽는법"

```python
data=["hello\n","hi\n","bye\n"]
```

```python
with open("0901_4.txt","w") as f:
    f.writelines(data)          # writelines은 여러줄에 걸쳐서 저장하는것
```

```python
with open("0901_4.txt","r") as f:
    print(f.readlines())          # readlines는 전체 데이터에 대해서 줄 단위로 리스트 요소로 저장된다.
                                  # \n까지 포함되어 저장된다는 것을 결과를 통해서 할 수 있다
```

```python
with open("0901_4.txt","r") as f:
    print(f.read())
```

```python
with open("0901_4.txt","r") as f:
    print(f.readline())            # readline 첫번째 줄만 읽는 방법
```

# 반복문 이용해서 읽는 방법

```python
with open("0901_4.txt","r") as f:
    print(f.readline())
```

```python
li=None # 내용이 아무것도 없을때 None을 사용  line이라는 변수를 사용하겠는데 line을 사용하고 싶을때
with open("0901_4.txt","r") as f:
    while li !="":                  #------> " "가운데 공백시 무한루프 발생
        li=f.readline()
        print(li.strip("\n"))
```

# return문:  함수를 호출한 곳으로 값을 전달하는 명령문

```python
def add2(x,y):
    return(x+y)
```

```python
print(add2(1,2))
```

```python
print(add2(y=10,x=20))       # x에 10 y에 20이 지정되나 변경할 수 있다.
                            # 인수를 지정해서 데이터값을 변경 할 수 있다.
```

```python
def add3(*arg):  # 인수명 앞에 *은 이 인수는 데이터를 하나 둘 열개든 전달되는 데이터가 상관이 없다.
    s=0
    for i in arg:
        s+=i
       # print(i)
  #  print(arg)
    return s
```

```python
#add3()         # 1. arg로 인해서 ()안에 데이터가 없어도 실행하는데 상관이 없다
add3(10)   # 2.() 즉 튜플 구조로 실행이됨
```

```python
#add3(10,20,30,40)
add3(10,20,30,40)
```

```python
def add4(op, *arg):
    res=1
    res2=999
    res3=333
    if op=="mul":
      #  return arg[0]*arg[1]*arg[2]
        for i in arg:
            res=res*i
    return res,res2,res3
```

```python
add4("mul",1,2,3)
```
