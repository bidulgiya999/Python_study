```python
re.match("[가-힣]","홍길동")
```

```python
re.match("[A-Z]+","Hello")
```


re.match 함수는 주어진 정규표현식으로 문자열의 처음부터 매치를 찾으며, 두 패턴 "[^A-Z]+"와 "^[A-Z]+" 사이에는 중요한 차이가 있습니다.

re.match("[^A-Z]+","Hello"):

이 패턴은 문자열 "Hello"의 처음부터 시작하여 대문자(A-Z)가 아닌 문자를 찾습니다. [^A-Z]+는 대문자가 아닌 문자를 하나 이상 찾으려고 시도합니다. "Hello"의 첫 번째 문자 "H"가 대문자이기 때문에 이 패턴과 일치하지 않습니다. 따라서 None을 반환합니다.

re.match("^[A-Z]+","Hello"):

이 패턴은 문자열 "Hello"의 처음부터 시작하여 대문자(A-Z)로 이루어진 문자열을 찾습니다. ^[A-Z]+는 대문자로 시작하는 문자열을 하나 이상 찾으려고 시도합니다. "Hello"의 첫 번째 문자 "H"가 대문자이므로 이 패턴과 일치합니다. 따라서 "H"를 반환합니다.

요약하면, 첫 번째 패턴은 대문자가 아닌 문자를 찾고, 두 번째 패턴은 대문자로 시작하는 문자


```python
re.match("[^A-Z]+","Hello")   # 매치안됨
```

```python
re.match("^[A-Z]+","Hello")    # 대문자로 시작한다
```

```python
re.match("^[A-Z]","Hello")
```

# group

### 123이 이름 45가 나이로 표현되고 싶을때

```python
res=re.match("([0-9]+) ([0-9]+)","123 45")  # 여기서 () 소괄호는 그룹의 의미를 표현
```

```python
res.group()  # 매칭된것 전체 추출하는 경우
res.group(1)
```

```python
res.group(2)
```

```python
res.groups()            #  튜플 단위로 출력됨
```

 문자열 : "print(123)"
 문자열에서 함수명과 인수를 각각의 그룹으로 나누어 출력
 res.group(0) : print
 res.group(1) : 123

```python
res2=re.search("[a-z1-9()]+","print(123)")                #  잘못된 것이므로 무시하면됨
res2.group(0)
```

```python
res=re.match("([a-z]+)[(]([0-9]+)[)]","print(123)")
print(res.group(1))
print(res.group(2))
```

```python
res4=re.match("([a-z]+)[(]([1-9]+)[)]","print(123)")
print(res4.group(1))
print(res4.group(2))
```
