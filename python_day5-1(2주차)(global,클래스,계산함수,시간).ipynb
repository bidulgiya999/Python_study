### global ( 전역변수와 지역변수에 대해 공부하기)

```python
res2 = 0
def add2(num):
    global res2 # 전역변수
    res2 += num
    return res2
```

```python
print(add2(1))
```

```python
res3=0
res4=0
def add3(num):
    global res3
    res3 += num
    return res3
def add4(num):
    global res4
    res4 += num
    return res4
```

```python
print(add3(3))   #둘이 더해서 누적합 출력을 원한다면
print(add3(2))   #5가 출력되기를 원한다면
```

```python
add3(1)
```

```python
add4(10)
```

# 클래스( 진짜 클래스 제대로 공부하기)

***__init__는 클래스로 부터 객체를 만드는 과정을 __init__이라고한다***

```python
class 붕어빵기계:
    def __init__(self,내용물):
        self.틀=내용물
    def 굽는다(self,재료):
        self.res+=재료
        return self.틀
```

```python
#붕1=붕어빵기계()
붕2=붕어빵기계("단팥")
붕2.틀
```

```python
class 사람:
    def __init__(self,n):
        print("사람이 태어 날때(객체 생성) 자동으로 호출됨")
        self.이름=n
```

```python
class 동물:
    def __init__(self,sp):   # self는 객체 자기 자신
        print("동물이 태어 날때(객체 생성) 자동으로 호출됨")
        self.종류=sp

    def 먹는다(self,f):
        self.주식=f


    def 공부한다(self,sub):
        print(sub,"를 공부한다")
        self.분야=sub
```

```python
#동물("파충류")
p1=동물("사람")
c1=동물("고양이")
```

```python
c1.종류
c1.공부한다("국어")
```

```python
p1.종류
p1.공부한다("딥러닝")
p1.분야
```
